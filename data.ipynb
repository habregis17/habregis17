```python
from google.colab import files
import pandas as pd
from re import sub
import io, os
uploaded = files.upload()
#variables
path_to_json='./'

# read file
for filename in os.listdir(path_to_json):
    if filename.endswith('.xlsx'):
        path = os.path.join(path_to_json, filename)
        df = pd.read_excel(path)
        os.remove(path)
df.columns = df.columns.str.replace(' ','')

df['EmplId']= df['EmplId'].values.astype(str)

data=df
count=df.shape[0]
my_dict = {'id':[],"firstName":[],"middleName":[],"lastName":[],'email':[],'telephone':[],'gender':[],'baseSalary':[],'organization':[],'doneBy':[],'emplId':[],'elementName':[],'elementDesc':[],'amount':[],'created_at':[],'updated_at':[]};
# names =noneDuplicate
for u in range(0,count):
  my_dict["id"].append(0)
  my_dict["email"].append('')
  my_dict["telephone"].append('')
  my_dict["gender"].append(0)
  my_dict["baseSalary"].append(0)
  my_dict["organization"].append('')
  my_dict["doneBy"].append('')
  my_dict["created_at"].append('2022-01-01')
  my_dict["updated_at"].append('2022-01-01')
  
  for i in range(0, len(df.columns)):

    if df.columns[i] =="Name":
      names=data[df.columns[i]].values.tolist()
      splited = names[u].split()

      if len(splited) ==2 :
        my_dict["firstName"].append(splited[0][:-1])
        my_dict["middleName"].append('')
        my_dict["lastName"].append(splited[1])
      else:
        my_dict["firstName"].append(splited[0][:-1])
        my_dict["middleName"].append(splited[1])
        my_dict["lastName"].append(splited[2])

    if df.columns[i] =="ElementName":
      elements=data[df.columns[i]].values.tolist()
      my_dict["elementName"].append(elements[u])

    if df.columns[i] =="ElementDesc":
      elementDesc=data[df.columns[i]].values.tolist()
      my_dict["elementDesc"].append(elementDesc[u])

    if df.columns[i] =="Payment":
      payment=data[df.columns[i]].values.tolist()
      my_dict["amount"].append(payment[u])

    if df.columns[i] =="EmplId":
      emplId=data[df.columns[i]].values.tolist()
      my_dict["emplId"].append(emplId[u])
      
newHeaders=pd.DataFrame.from_dict(my_dict)    
#saving data
newHeaders.to_csv("structured_data.csv",index=False)
files.download('structured_data.csv')
```
